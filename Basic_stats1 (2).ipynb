```python
#standarization of numerical values
```

```python
import pandas as pd

# Load the dataset from a CSV file
df = pd.read_csv('/content/sales_data_with_discounts.csv')

# Selecting only the numerical columns
numerical_columns = df.select_dtypes(include=['number'])

# Standardizing the numerical columns
standardized_columns = (numerical_columns - numerical_columns.mean()) / numerical_columns.std()

# Replacing the original numerical columns with the standardized ones
df[numerical_columns.columns] = standardized_columns

# Save the standardized DataFrame to a new CSV file
df.to_csv('standardized_file.csv', index=False)

# Print the standardized DataFrame
print(df)
```
