State the Hypotheses statement

```python
def state_hypotheses():

    null_hypothesis = (
        "H₀: The mean weekly operating cost is equal to the theoretical mean operating cost "
        "(μ = 4,000)."
    )


    alternative_hypothesis = (
        "H₁: The mean weekly operating cost is greater than the theoretical mean operating cost "
        "(μ > 4,000)."
    )


    print("Hypotheses for Investigating Weekly Operating Costs:")
    print(null_hypothesis)
    print(alternative_hypothesis)


state_hypotheses()
```

Calculate the Test Statistic

```python
import numpy as np

# Given data
x_bar = 3050  # Sample mean weekly cost
mu = 4000     # Theoretical mean weekly cost
sigma = 5 * 25   # Standard deviation (5 * 25)
n = 25        # Sample size

# Step 2: Calculate the test statistic
t_statistic = (x_bar - mu) / (sigma / np.sqrt(n))
print(f"Test Statistic (t): {t_statistic}")
```

Determine the Critical Value

```python
import scipy.stats as stats

alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha)
print(f"Critical Value (Z): {critical_value}")
```

Make a Decision

```python
if t_statistic > critical_value:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

print(f"Decision: {decision}")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency, chi2
```

```python
# 1. State the Hypotheses
print("Hypotheses:")
print("H0: There is no association between device type and customer satisfaction.")
print("H1: There is a significant association between device type and customer satisfaction.\n")
```

```python
# 2. Create the contingency table
data = np.array([
    [50, 70],    # Very Satisfied
    [80, 100],   # Satisfied
    [60, 90],    # Neutral
    [30, 50],    # Unsatisfied
    [20, 50]     # Very Unsatisfied
])

device_labels = ["Smart Thermostat", "Smart Light"]
satisfaction_levels = [
    "Very Satisfied", "Satisfied", "Neutral", "Unsatisfied", "Very Unsatisfied"
]

df = pd.DataFrame(data, columns=device_labels, index=satisfaction_levels)
print("Contingency Table:\n", df, "\n")
```

```python
# 3. Compute the Chi-Square Statistic
chi2_stat, p_val, dof, expected = chi2_contingency(data)

print(f"Chi-Square Statistic = {chi2_stat:.4f}")
print(f"Degrees of Freedom = {dof}")
print("Expected Frequencies Table:\n", pd.DataFrame(expected, columns=device_labels, index=satisfaction_levels), "\n")
```

```python
# 4. Determine the Critical Value
alpha = 0.05
critical_value = chi2.ppf(q=1 - alpha, df=dof)
print(f"Critical Value at alpha=0.05: {critical_value:.4f}")
```

```python
# 5. Make a Decision
print("\nDecision:")
if chi2_stat > critical_value:
    print("Since Chi-Square statistic > Critical Value, we reject the null hypothesis.")
    print("Conclusion: There is a significant association between device type and customer satisfaction.")
else:
    print("Since Chi-Square statistic <= Critical Value, we fail to reject the null hypothesis.")
    print("Conclusion: There is no significant association between device type and customer satisfaction.")
```
